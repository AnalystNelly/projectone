```python
import pandas as pd
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/databricks/Spark-The-Definitive-Guide/master/data/retail-data/all/online-retail-dataset.csv')
df
```

```python
df.info()
```

```python
df[df['CustomerID'].notnull()].reset_index()
```

```python
df = df[df['CustomerID'].notnull()].reset_index(drop=True)
df
```

```python
df.info()
```

```python
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])
```

```python
df['CustomerID'] = df['CustomerID'].astype(int).astype(str)
```

```python
df.info()
```

```python
df.describe()
```

```python
df = df[df['Quantity']>0].reset_index(drop=True)
```

```python
df.describe()
```

```python
df['TotalPrice'] = df['Quantity'] * df['UnitPrice']
```

```python
df.head()
```

```python
#Recency
df.groupby('CustomerID')['InvoiceDate'].last()
```

```python
#frequency
df.groupby('CustomerID')['InvoiceNo'].count()
```

```python
#
df.groupby('CustomerID')['TotalPrice'].sum()
```

```python
data = df.groupby('CustomerID').agg({
    'InvoiceDate' : 'last',
    'CustomerID' : 'count',
    'TotalPrice' : 'sum'
})

data
```

```python
most_recent_date = df['InvoiceDate'].max()
most_recent_date
```

```python
data['DurationLastPurchase'] = (most_recent_date - data['InvoiceDate']).dt.days
```

```python
data.head()
```

```python
data.rename(columns = {
    'CustomerID' : 'F',
    'TotalPrice' : 'M',
    'DurationLastPurchase' : 'R'
}, inplace=True)
```

```python
data = data.reset_index()
```

```python
data = data[['CustomerID', 'F','M','R']]
```

```python
data
```

```python
data['R'] = 1-data['R']/max(data['R'])
data['M'] = data['M']/max(data['M'])
data['F'] = data['F']/max(data['F'])
```

```python
data
```

```python
data['RFM'] = 0.15*data['R'] + 0.28*data['F'] + 0.57*data['M']
```

```python
data
```

```python
data['RFM'].plot(kind='hist')
```

```python
data[data['RFM']>0.15]
```

```python
data.sort_values(by='RFM', ascending=False).head(25)
```

```python
data.sort_values(by='RFM',ascending=False).head(25)['RFM'].plot(kind='bar')
```
